$\bf 5)$ Synthesize a wave by creating a spectrum with arbitrary harmonics, invoking make_wave, and listening. What happens as you add frequency components that are not multiples of the fundamental?

We initially declare two signals, one cosine and one sine signal of different frequencies and of amplitude equal to 1.

```python
cos_signal = thinkdsp.CosSignal(freq = 440, amp=1.0, offset = 0)
sin_signal = thinkdsp.SinSignal(freq = 880, amp=1.0, offset = 0)
```

We are going to now add the two signals and invoke $make\_wave()$ from the thinkdsp library.

```python
# Adds the two signals
mix_signal = cos_signal + sin_signal

# Makes signal to wave with make_wave() method
wave_signal = mix_signal.make_wave(duration = 3, start = 0, framerate = 11025)
```

We can listen to that signal with the following code.

```python
# Plays audio
wave_signal.make_audio()
```

We can also plot $wave\_signal$ and see how it looks like for three periods through the following code.

```python
# Grabs the period of the signal
period = mix_signal.period

# Makes wave with three periods
wave_signalNew = mix_signal.make_wave(duration = period*3, start = 0, framerate = 11025)

# Plots wave_signal graph
wave_signalNew.plot()
```

Plotting three periods of $wave\_signal$ one can easily see the periodicity of the signal. Let's now add frequencies that are multiples of the fundamental and see what happens.

```python
# Harmonic integer multiple of fundamental
harm1 = thinkdsp.SinSignal(freq = 1320, amp=1.0, offset = 0)

# Added to mix_signal
mix_signal += harm1

# New wave definition with harm1 signal
wave_signalHarm1 = mix_signal.make_wave(duration = 3, start = 0, framerate = 11025)

# PLay audio
wave_signalHarm1.make_audio()
```

By adding this additional integer multiple of the fundamental frequency, we generated a sound that is pleasant to hear. What if now we add a frequency that is not a multiple of the fundamental? The following happens. 

```python
# Harmonic non-integer multiple of fundamental
harm2 = thinkdsp.SinSignal(freq = 1600, amp=1.0, offset = 0)

# Added to mix_signal
mix_signal += harm2

# New wave definition with harm1 signal
wave_signalHarm2 = mix_signal.make_wave(duration = 3, start = 0, framerate = 11025)

# PLay audio
wave_signalHarm2.make_audio()
```

The addition of $harm2$ resulted in a very bad sound as the harmonic spectrum of the signal now contains frequencies that are not only integer multiples of the fundamental frequency.

$\bf 6)$ A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats. Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.  Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

The $SawtoothSignal$ class that extends Signal and provides an evaluate method - $evaluate()$ - to evaluate the y's of the wave is given below.

```python
import numpy
import math

class SawtoothSignal(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        cycles = self.freq*ts + self.offset/(math.pi*2)
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
```

We can even plot the wave of that signal to make sure it has the correct structure, as shown bellow.

```python
sawtooth_signal = SawtoothSignal(440)
sawtooth_segmentExample = sawtooth_signal.make_wave(duration=sawtooth_signal.period*3, start = 0, framerate=10000)
sawtooth_segmentExample.plot()
```

$\bf 9)$ Compute the spectrum of an 1100 Hz square wave sampled at 10 kHz, and compare it to the spectrum of a triangle wave in Figure 2.5.

```python
square_signal_new = thinkdsp.SquareSignal(freq=1100)

square_segment_new = square_signal_new.make_wave(duration=3, start = 0, framerate=10000)
square_spectrum_new = square_segment_new.make_spectrum()
square_spectrum_new.plot()
```

The graph above and the one depicted in Figure 2.5. of the book ThinkDSP by Allen Downey are both representing aliasing. However, it is important to note that in the case of the SquareSignal, the aliasing is amplified, as there are more frequencies that are aliased and that the amplitude is greater and also there is much more "noise" in frequencies that are not integer multiples of the fundamental frequency. The reason the SquareSignal has aliased frequencies with higher amplitudes is that the harmonics of the square drop off in proportion to $\frac{1}{f}$, while the harmonics of the triangle wave drop off like $\frac{1}{f^2}$. Square waves have more frequency components than triangles that fold back and result in aliasing.

$\bf 10)$ The triangle and square waves have odd harmonics only; the sawtooth wave has both even and odd harmonics. The harmonics of the square and sawtooth waves drop off in proportion to 1/f; the harmonics of the triangle wave drop off like 1/f2. Can you find a waveform that has even and odd harmonics that drop off like $1/f^2$?

We are going to initiate an empty numpy array that we are going to populate with amplitude values based on frequency given. We are, therefore, creating a waveform that has both even and odd harmonics, that drop off like $1/f^2$.

```python
# Initiate empty array of 5000 spaces where the data types are complex numbers
hs = numpy.zeros(5000, dtype=numpy.complex64)

# Initiate 'for' loop for frequencies
for freq in range(100,5000,100):
    # Assign new values to amplitudes     
    hs[freq] = 1/(freq**2)
    
# Create Spectrum
my_spec = thinkdsp.Spectrum(hs,framerate=10000)

# Plot my_spec graph
my_spec.plot()
```

```python
wave_spec = my_spec.make_wave()
wave_spec.segment(start = 0, duration = 0.02).plot()
```
