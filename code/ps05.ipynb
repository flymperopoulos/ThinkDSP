$\bf 1)$ Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes. The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy as np
import scipy.signal

%precision 3
%matplotlib inline

PI2 = 2 * math.pi
i = complex(0, 1)
```

We can use a method that approximates if things are equal.

```python
def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol
```

The question checks to see if there is a difference between $\it a$ and $\it b$ and returns a boolean depending on the result.

```python
N = 7

# defines the time array from 0 to (N-1/N) 
ts = numpy.arange(N) / N

# first array of frequencies from 0 to (N-1)
freqs = numpy.arange(N)

# second array of frequencies from N to (2N-1)
freqs2 = numpy.arange(N, 2*N)

# Respective products taking place
args = numpy.outer(ts, freqs)
args2 = numpy.outer(ts,freqs2)

# Final products
M = numpy.exp(i * PI2 * args)
M2 = numpy.exp(i * PI2 * args2)
amps = [0, 1, 2, 3, 4, 5, 6, 7]

# Compares them and generateas array of booleans
print(approx_equal(M,M2))
```

As seen above, it is shown that the two cases yield almost identical results, as each comparison returns a boolean $\it True$. 

$\bf 2)$ In thinkdsp.py, the Spectrum methods $low\_pass$ and $high\_pass$ are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want). Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

We can simply implement a low pass filter that has a cut-off frequency at 40Hz. To do that, we will create an array of ones (1's) 40 times and then after those values, we will simetrically create another 40 zeros (0's) for the all the frequencies that are equal to zero and above 440Hz.

```python
# passing frequencies
pass_freqs = np.ones(40)

# non-passing frequencies
non_pass_freqs = np.zeros(40)

# generating signal
low_pass_signal = np.hstack((pass_freqs, non_pass_freqs))

# plots the resulting signal
thinkplot.plot(low_pass_signal)
```

```python
res_low_wave = thinkdsp.Spectrum(low_pass_signal, 44100).make_wave()

res_low_wave.plot()
```

Similarly, we can now create a $high\_pass$ with the same cut-off frequency at 40Hz.

```python
# passing frequencies
pass_freqs = np.zeros(40)

# non-passing frequencies
non_pass_freqs = np.ones(40)

# generating signal
high_pass_signal = np.hstack((pass_freqs, non_pass_freqs))

# plots the resulting signal
thinkplot.plot(high_pass_signal)
```

```python
res_high_wave = thinkdsp.Spectrum(high_pass_signal, 44100).make_wave()

res_high_wave.plot()
```

The above results indicate the windows for a low pass and a high pass filter. It is significant to point out that both resulting waves, express "abrupt" behavior with peaks being demonstrated at the high positive, respectively, values.

$\bf 4)$ Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this window is an increasing function of f, so you can think of diff as a high-pass filter. One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f. If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like. Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window? See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

window = np.array([1.0, -1.0])
window = zero_pad(window,500)
wave = thinkdsp.Wave(window, framerate=10000)
wave.make_spectrum().plot()
```

```python
window2 = np.array([-1/2,2,-3/2])
```

```python
window2 = zero_pad(window2,500)
wave2 = thinkdsp.Wave(window2, framerate=10000)
wave2.make_spectrum().plot()
```

```python
window3 = np.array([1, -2, 1])
window3 = zero_pad(window,500)
wave3 = thinkdsp.Wave(window, framerate=10000)
wave3.make_spectrum().plot()
```

Looking at all the above in one graph will help us develop a conclusion about the relationships between the explored situations above.

```python
wave.make_spectrum().plot()
wave2.make_spectrum().plot()
wave3.make_spectrum().plot()
```

The filters above yield several interesting results about the derivatives of the graphs produce. Similarities are shown and basically, the first derivative is a high pass filter, like filtering with a column vector [1,-1] across a square wave.

$\bf 6)$ Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
signal = thinkdsp.ParabolicSignal(freq=440)
parabolic_wave = signal.make_wave(duration=signal.period*4, framerate=44100)
parabolic_wave.plot()
```

```python
parabolic_spectrum = parabolic_wave.make_spectrum()
parabolic_spectrum.plot()
```

```python
parabolic_spectrum.differentiate()
parabolic_spectrum.plot()
```

It can be seen that when we apply the method $differentiate()$ to the generated spectrum for the parabolic signal, we get a graph that indicates the increasing value in the calculated derivated respective amplitude.
